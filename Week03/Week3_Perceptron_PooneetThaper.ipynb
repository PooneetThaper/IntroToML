```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
from matplotlib.colors import ListedColormap
```

```python
class Perceptron01(object):
    """
    Perceptron classifier (target values in {0,1})
    
    Code from "Python Machine Learning," Sebastian Raschka, Packt Publishing, 2015, with modifications.
    
    Parameters
    ------------
    eta : float
        Learning rate (between 0.0 and 1.0)
    n_iter : int
        Number of passes over the training set
        
    Attributes    
    ------------
    w_ : id-array
        Weights after fitting
    errors: list
        Number of misclassifications in each epoch        
    """    
    def __init__(self, eta=0.2, n_iter=10):
        self.eta = eta
        self.n_iter = n_iter
    def fit(self, X, y):
        """
        Fit training data
        
        Paramters
        ------------
        X: {array-like}, shape = [n_samples, n_features]
            Training vectors, where n_samples is the number of samples and
            n_features is the number of features.
        y : array-like, shape = [n_samples]
            Target values
            
        Returns
        ------------
        self : object
        
        """
        self.w_ = np.zeros(1 + X.shape[1])
        self.weights_ = []
        self.errors_ = []        
        for i in range(self.n_iter):
            errors = 0
            for xi, target in zip(X,y):
                update = self.eta * (target - self.predict(xi))
                self.w_[1:] += update*xi
                self.w_ [0] += update
                errors += int(update != 0.0)
            self.errors_.append(errors)
            self.weights_.append(self.w_.tolist())
            i += 1
        return self
    
    def net_input(self, X):
        """ Calculate net input """
        return np.dot(X, self.w_[1:]) + self.w_[0]
    
    def predict(self, X):
        """ Return class label after each setp """
        return np.where(self.net_input(X) >= 0.0, 1, 0)
```

```python
# get Iris classifications

IC = np.unique(IA[:,4]) # Unique classes (order same each time because data order)
pltC = ('red', 'blue', 'green') # Color for plotting each class
pltM = ('o', 'x', '*') # Marker for plotting each class
for i in range(0,3):
    print(IC[i], pltC[i], pltM[i])
    
```

```python
# Setup Iris feature lablels

IrisF=("sepal length in cm (0)", "sepal width in cm (1)", "petal length in cm (2)", "petal width in cm (3)")
for i in range(0,4):
    print(IrisF[i])
```

```python
# plot classifications of all pairs of features 

plt.close('all')
fig, ax = plt.subplots(nrows=3, ncols=2, figsize= (8,12), sharex=True, sharey=True)

Cl = IA[:,4] #Classifications from the data
i = 0 # Y axis feature number
k = 0 # Plot number
k1 = 0 # Row number of plot
k2 = 0 # Column number of plot
while i < 4:
    yF = IrisF[i] # Feature on y axis
    j = 0 # X axis feature number
    while j < i:
        k1 = np.int(k/2) # Row of this plot
        k2 = np.mod(k,2) # Column of this plot
        xF = IrisF[j] # Feature on x axis
        ax[k1,k2].scatter(x=IA[Cl == IC[0], j], y=IA[Cl == IC[0], i], color=pltC[0], marker=pltM[0], label=IC[0])
        ax[k1,k2].scatter(x=IA[Cl == IC[1], j], y=IA[Cl == IC[1], i], color=pltC[1], marker=pltM[1], label=IC[1])
        ax[k1,k2].scatter(x=IA[Cl == IC[2], j], y=IA[Cl == IC[2], i], color=pltC[2], marker=pltM[2], label=IC[2])
        ax[k1,k2].set(aspect='equal')
        ax[k1,k2].set_xlim([0,9])
        ax[k1,k2].set_ylim([0,9])
        ax[k1,k2].set_xlabel(xF)
        ax[k1,k2].set_ylabel(yF)
        ax[k1,k2].legend(loc='upper left', fontsize='x-small')
        j += 1
        k += 1
    i += 1
plt.show()    
```

```python
# Separable example:
# focus on classification of "setosa" and "versicolor", class indexes 0 and 1
cdx = np.array([0,1]) # Class indexes
# focus on the features "sepal length" and "petal length", feature indexes 0 and 2
fdx = np.array([0,2]) # Feature indexes
```

```python
# Building Dataset
i = 0
numCl = 0 # Number of samples of selected classes (100 in this case)
for i in range(0 , len(Cl)):
    if Cl[i] == IC[cdx[0]]:
        numCl += 1
    if Cl[i] == IC[cdx[1]]:
        numCl += 1    
    i += 1  

NC = np.zeros((numCl)) # Classes for classification
NF = np.zeros((numCl, 2)) # Features for classification

i = 0 # Counter for iterating Cl (array of classes for all samples in dataset)
j = 0 # Counter for iterating position in NC and NF
for i in range(0 , len(Cl)):
    # if class of sample is one of selected classes:
    #     set next position in NC and NF to sample class and its features, respectively
    #     iterate to next position in NC and NF
    # iterate down Cl
    if Cl[i] == IC[cdx[0]]:
        NC[j] = 0
        NF[j,0] = IA[i,fdx[0]]
        NF[j,1] = IA[i,fdx[1]]
        j += 1
    if Cl[i] == IC[cdx[1]]:      
        NC[j] = 1
        NF[j,0] = IA[i,fdx[0]]
        NF[j,1] = IA[i,fdx[1]]
        j += 1
    i += 1

plt.close('all')
plt.scatter(x = NF[NC == 0, 0], y = NF[NC == 0, 1], color=pltC[cdx[0]], marker=pltM[cdx[0]], label=IC[cdx[0]])
plt.scatter(x = NF[NC == 1, 0], y = NF[NC == 1, 1], color=pltC[cdx[1]], marker=pltM[cdx[1]], label=IC[cdx[1]])
plt.xlabel(IrisF[fdx[0]])
plt.ylabel(IrisF[fdx[1]])
plt.legend(loc='lower right', fontsize='medium')
plt.show()
```

```python
plt.close('all')
def plot_decision_regions(NF,NC, classifier, resolution=0.02):
    colors = ('red', 'blue')
    cmap = ListedColormap(colors[:len(np.unique(NC))])
    x1_min, x1_max = NF[:, 0].min() - 1, NF[:, 0].max() + 1
    x2_min, x2_max = NF[:, 1].min() - 1, NF[:, 1].max() + 1
    xx1, xx2 = np.meshgrid(np.arange(x1_min, x1_max, resolution), np.arange(x2_min, x2_max, resolution))
    Z = classifier.predict(np.array([xx1.ravel(), xx2.ravel()]).T)
    Z = Z.reshape(xx1.shape)
    plt.contourf(xx1, xx2, Z, alpha=0.2, cmap=cmap)
    plt.xlim(xx1.min(), xx1.max())
    plt.ylim(xx2.min(), xx2.max()) 
    for idx, cls in enumerate(np.unique(NC)):
        plt.scatter(x=NF[NC == cls, 0], y=NF[NC == cls, 1], alpha=0.8, color=pltC[cdx[idx]], marker=pltM[cdx[idx]], label=IC[cdx[idx]])
plot_decision_regions(NF, NC, classifier=ppn)
plt.xlabel(IrisF[fdx[0]])
plt.ylabel(IrisF[fdx[1]])
plt.legend(loc='upper left', fontsize='medium')
plt.show()
```

```python
# Building Dataset
i = 0
numCl = 0 # Number of samples of selected classes (100 in this case)
for i in range(0 , len(Cl)):
    # if class of sample i in data is one of selected:
    #     add one to number of samples
    # increment iterator for samples' class array
    if Cl[i] == IC[cdx[0]]:
        numCl += 1
    if Cl[i] == IC[cdx[1]]:
        numCl += 1    
    i += 1  

# Dataset
NC = np.zeros((numCl)) # Classes for classification
NF = np.zeros((numCl, 2)) # Features for classification
    
i = 0 # Counter for iterating Cl (array of classes for all samples in dataset)
j = 0 # Counter for iterating position in NC and NF
for i in range(0 , len(Cl)):
    # if class of sample is one of selected classes:
    #     set next position in NC and NF to sample class and its features, respectively
    #     iterate to next position in NC and NF
    # iterate down Cl
    if Cl[i] == IC[cdx[0]]:
        NC[j] = 0
        NF[j,0] = IA[i,fdx[0]]
        NF[j,1] = IA[i,fdx[1]]
        j += 1
    if Cl[i] == IC[cdx[1]]:      
        NC[j] = 1
        NF[j,0] = IA[i,fdx[0]]
        NF[j,1] = IA[i,fdx[1]]
        j += 1
    i += 1
    
def shuffle_together(a, b):
    assert(len(a)==len(b))
    new_a = np.zeros_like(a)
    new_b = np.zeros_like(b)
    permutation = np.random.permutation(len(a))
    for old_index, new_index in enumerate(permutation):
        new_a[new_index] = a[old_index]
        new_b[new_index] = b[old_index]
    return new_a, new_b

NC, NF = shuffle_together(NC, NF)

plt.close('all') #Close plot
# Scatter values of features samples with first selected class (NC==0) with first selected feature on X axis
plt.scatter(x = NF[NC == 0, 0], y = NF[NC == 0, 1], color=pltC[cdx[0]], marker=pltM[cdx[0]], label=IC[cdx[0]])
# Scatter values of features samples with second selected class (NC==1) with first selected feature on X axis
plt.scatter(x = NF[NC == 1, 0], y = NF[NC == 1, 1], color=pltC[cdx[1]], marker=pltM[cdx[1]], label=IC[cdx[1]])
# Label each axis with the applicable feature
plt.xlabel(IrisF[fdx[0]])
plt.ylabel(IrisF[fdx[1]])
# Add a legend (and position out of the way)
plt.legend(loc='right', fontsize='medium')
# Show plot inline
plt.show()
```
